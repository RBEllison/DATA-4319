```julia
function sign(w, x)
    x = [x[1], x[2], 1.0]
    return w'x > 0 ? 1 : -1
end
```

```julia
function perpetron_update(w, x, y)
    if sign(w, x) != y
        w += y*[x[1], x[2], 1.0]
    end
    return w
end
```

```julia
function error(w, feature_set, labels)
    return sum([sign(w, feature_set[i]) != labels[i] ? 1 : 0 for i =  1:length(feature_set)])
end
```

```julia
function perceptron(w, feature_set, labels, e)
    errors = []
    append!(errors, error(w, feature_set, labels))
    
    weights = [w]
    
    while error(w, feature_set, labels) > 0
        for i = 1:length(feature_set)
            w = perpetron_update(w, feature_set[i], labels[i])
        end
        
        append!(weights, w)
        append!(errors, error(w, feature_set, labels))
    end
    
    return w, weights, errors
end
```
